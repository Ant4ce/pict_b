```python
stk_names_financials = []
stks_yf_fin = []
```

```python
def download_stk_financials():
    print("starting now!")
    for steps in range(1, 12):
        print("current step: {}".format(steps))
        for stock in nasdaq_list["STOCKS"][500 *(steps - 1): 500 * steps]:
            current_stk = yf.Ticker(stock)
            print("current stock: {}".format(stock))
            if len(current_stk.financials) == 22:
                print("added: {}".format(stock))
                stk_names_financials.append(stock)
                stks_yf_fin.append(current_stk.financials)
        
        print("going to sleep")
        time.sleep(3600)
        print("finished sleeping, resuming!")
```
